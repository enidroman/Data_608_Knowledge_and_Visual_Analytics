 Population dataset came from https://worldpopulationreview.com/states

```python
# Load the dataset after is saved in github.
df1 = pd.read_csv('https://raw.githubusercontent.com/enidroman/Data_608_Knowledge_and_Visual_Analytics/main/IIJA%20FUNDING%20AS%20OF%20MARCH%202023(1)%20(1).csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/enidroman/Data_608_Knowledge_and_Visual_Analytics/main/state-population.csv')
df3 = pd.read_csv('https://raw.githubusercontent.com/enidroman/Data_608_Knowledge_and_Visual_Analytics/ba6668f687ce482e2ac1cd0ba897f036bebbaaec/2020-president.csv')
```

### **Infrastructure Investment & Jobs Act Funding Allocation by State**

```python
df1 = df1.rename(columns={'State, Teritory or Tribal Nation': 'State'})
df1.head()
```

### **Infrastructure Investment & Jobs Act Funding Allocation and Population by State**

### **Population by State**

```python
import matplotlib.pyplot as plt

sort_df = df_merged_all.sort_values(by='Pop 2023', ascending=False)

# Create the bar chart
plt.figure(figsize=(17, 14))
bars = plt.bar('State', 'Pop 2023', data=sort_df, color='blue')
plt.xlabel('State')
plt.ylabel('Population')
plt.title('Population by State')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability


# Add annotations above the bars
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.2f}',  # Format the annotation text as needed
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 10),  # Adjust the vertical position
                 textcoords="offset points",
                 ha='center',
                 fontsize=10,
                 rotation=90)


# Show the chart
plt.tight_layout()
plt.show()
```

### **Infrastructure Investment & Jobs Act Funding Allocation by State**

```python
import matplotlib.pyplot as plt

sort_df = df_merged_all.sort_values(by='Total (Billions)', ascending=False)

# Create the bar chart
plt.figure(figsize=(17, 14))
bars = plt.bar('State', 'Total (Billions)', data=sort_df, color='blue')
plt.xlabel('State')
plt.ylabel('Total (Billions)')
plt.title('Allocation of the Infrastructure Investment and Jobs Act funding by State and Territory')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability


# Add annotations above the bars
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.2f}',  # Format the annotation text as needed
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 10),  # Adjust the vertical position
                 textcoords="offset points",
                 ha='center',
                 fontsize=10,
                 rotation=90)


# Show the chart
plt.tight_layout()
plt.show()
```

###**Political Party by State**

```python

# Extract the relevant columns for plotting
states = df_merged_all['State']
party = ['Democrat', 'Republican', 'Libertarian', 'Other']

# Define the width of each bar and the x-axis positions
bar_width = 0.20
index = np.arange(len(states))

# Create a color map for parties
party_colors = {
    'Democrat': 'blue',
    'Republican': 'red',
    'Libertarian': 'green',
    'Other': 'purple'
}

# Create the side-by-side grouped bar chart
plt.figure(figsize=(17, 10))

for i, party in enumerate(party):
    plt.bar(index + i * bar_width, df_merged_all[party], bar_width, label=party, color=party_colors[party])

plt.xlabel('State')
plt.ylabel('Votes')
plt.title('Votes by State')
plt.xticks(index + bar_width * (len(party) - 1) / 5, states)
plt.xticks(rotation=90)


plt.legend()

plt.tight_layout()
plt.show()
```

#**Question 1**


# **Is the allocation equitable based on the population of each of the States and Territories, or is bias apparent?**

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=df_merged_all['State'], y=df_merged_all['Total (Billions)'], name="Total (Billions)"),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=df_merged_all['State'], y=df_merged_all['Pop 2023'], name="Population"),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text="Allocation of the Infrastructure Investment and Jobs Act funding and Population by State and Territory"
)

# Set x-axis title
fig.update_xaxes(title_text="State")

# Set y-axes titles
fig.update_yaxes(title_text="Total (Billions)", secondary_y=False)
fig.update_yaxes(title_text="Population", secondary_y=True)

fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(rows=2, cols=1,
                    specs=[[{"secondary_y": True}], [{"secondary_y": True}]],
                    shared_xaxes=True  # Share the x-axis
                           )

# Top left
fig.add_trace(
    go.Scatter(x=df_merged_all['State'], y=df_merged_all['Total (Billions)'], name="Total Billions"),
    row=1, col=1,secondary_y=False)

fig.add_trace(
    go.Scatter(x=df_merged_all['State'], y=df_merged_all['Pop 2023'], name="Population"),
    row=2, col=1, secondary_y=False
)

# Customize the layout and add a title
fig.update_layout(title_text='Allocation vs. Population by State',
                  xaxis_tickangle=-45,
                                    )

fig.show()
```

As per the graph above, I don't see any bias in the allocation of funds that were distributed to each state. I believe the allocation of funds were equitable based on the population of each of the States and Territories.

# **Does the allocation favor the political interests of the Biden administration?****

```python
import matplotlib.pyplot as plt

# Create a figure with two subplots
fig, ax1 = plt.subplots(figsize=(16, 7))

# Create the bar chart on the left axis
ax1.bar(df_merged_all['State'], df_merged_all['Democrat'], color='b', alpha=0.7, label='Bar Chart')
ax1.set_xlabel('State')
ax1.set_ylabel('Democrat', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Rotate the x-axis labels for the bar chart
ax1.set_xticklabels(df_merged_all['State'], rotation=90)  # Rotate the x-axis labels

# Create a second y-axis on the right for the line chart
ax2 = ax1.twinx()
ax2.plot(df_merged_all['State'], df_merged_all['Total (Billions)'], color='y', marker='o', linestyle='-', label='Line Chart')
ax2.set_ylabel('Total (Billions)', color='y')
ax2.tick_params(axis='y', labelcolor='y')

# Add a legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2
ax1.legend(lines, labels, loc='upper left')

# Add a title
plt.title('Allocation of the Infrastructure Investment and Jobs Act funding by State and Territory - Democrat Party')

plt.tight_layout()
plt.show()
```
