```python
# Load the pandas package.
import pandas as pd
import numpy as np
```

## **THE DATA SET**

The dataset was taken by extracting the salary data by state and title role from https://www.salary.com/ by a method called scraping.

Scraping codes is shown on https://colab.research.google.com/drive/1APBq6vtVElKCXCLdAI_yknoOz9LUb5hA

```python
# Load the dataset after is saved in github.
df = pd.read_csv('https://raw.githubusercontent.com/enidroman/Data_608_Knowledge_and_Visual_Analytics/main/salary-state-results.csv')
df.head()
```

## **DATA CLEANING**

```python
# Renaming Columns.
# create a dictionary

dict = {'Location': 'State',
        'nTile10': '10th Percentile',
        'nTile25': '25th Percentile',
        'nTile50': '50th Percentile',
        'nTile75': '75th Percentile',
        'nTile90': '90th Percentile'}

# call rename () method
df.rename(columns=dict,
          inplace=True)
df.head()
```

```python
# Adding column Abbreviation.
us_state_to_abbrev = {
    "Alabama": "AL",
    "Alaska": "AK",
    "Arizona": "AZ",
    "Arkansas": "AR",
    "California": "CA",
    "Colorado": "CO",
    "Connecticut": "CT",
    "Delaware": "DE",
    "District of Columbia": "DC",
    "Florida": "FL",
    "Georgia": "GA",
    "Hawaii": "HI",
    "Idaho": "ID",
    "Illinois": "IL",
    "Indiana": "IN",
    "Iowa": "IA",
    "Kansas": "KS",
    "Kentucky": "KY",
    "Louisiana": "LA",
    "Maine": "ME",
    "Maryland": "MD",
    "Massachusetts": "MA",
    "Michigan": "MI",
    "Minnesota": "MN",
    "Mississippi": "MS",
    "Missouri": "MO",
    "Montana": "MT",
    "Nebraska": "NE",
    "Nevada": "NV",
    "New Hampshire": "NH",
    "New Jersey": "NJ",
    "New Mexico": "NM",
    "New York": "NY",
    "North Carolina": "NC",
    "North Dakota": "ND",
    "Ohio": "OH",
    "Oklahoma": "OK",
    "Oregon": "OR",
    "Pennsylvania": "PA",
    "Rhode Island": "RI",
    "South Carolina": "SC",
    "South Dakota": "SD",
    "Tennessee": "TN",
    "Texas": "TX",
    "Utah": "UT",
    "Vermont": "VT",
    "Virginia": "VA",
    "Washington": "WA",
    "West Virginia": "WV",
    "Wisconsin": "WI",
    "Wyoming": "WY",
    }

# Use the map function to add a new "Abbreviation" column
df["Abbreviation"] = df["State"].map(us_state_to_abbrev)


# Print the modified DataFrame
df.head()
```

```python
# Remove all columns between column index 1 to 5 and 6 and 8
df.drop(df.iloc[:, 2:5], inplace=True, axis=1),
df.drop(df.iloc[:, 3:5], inplace=True, axis=1)
df.head()
```

```python
# Rename 50th Percentile column to Median Salary
df = df.rename(columns={
    "50th Percentile": "Median Salary",
    "Abbreviation" : "State Abbr"
})
df.head()
```

```python
# Get unique values in the "Title" column
unique_titles = df["Title"].unique()

# Convert the unique titles to a list
unique_titles_list = unique_titles.tolist()

# Print the list of unique titles
print(unique_titles_list)
```

```python
# Moving Abbreviation Column.
df = df.reindex(columns=['Title', 'State', 'State Abbr', 'Median Salary'])
df
```

```python
# Data Types.
df.dtypes
```

```python
# Filter the DataFrame to include only "Data Scientist" records
data_scientist_df = df[df["Title"] == "Data Scientist"]

# Group the "Data Scientist" DataFrame by "State" and "Median Salary" and calculate the mean salary
grouped1 = data_scientist_df.groupby(["State Abbr"])["Median Salary"].mean().reset_index(name="Mean Salary")

# Print the grouped DataFrame
print(grouped1)
```

```python
# Filter the DataFrame to include only "Data Scientist" records
data_analyst_df = df[df["Title"] == "Data Analyst"]

# Group the "Data Scientist" DataFrame by "State" and "Median Salary" and calculate the mean salary
grouped2 = data_analyst_df.groupby(["State Abbr"])["Median Salary"].mean().reset_index(name="Mean Salary")

# Print the grouped DataFrame
print(grouped2)
```

```python
# Filter the DataFrame to include only "Data Scientist" records
data_business_analyst_df = df[df["Title"] == "Business Analyst"]

# Group the "Data Scientist" DataFrame by "State" and "Median Salary" and calculate the mean salary
grouped5 = data_business_analyst_df.groupby(["State Abbr"])["Median Salary"].mean().reset_index(name="Mean Salary")

# Print the grouped DataFrame
print(grouped5)
```

```python

import pandas as pd
import plotly.express as px


# Define custom colors for specific job titles
custom_colors = {
    "Business Analyst": "green",  # Custom color for Business Analyst
    "Data Analyst": "goldenrod",      # Custom color for Data Analyst
}

# Create an interactive line graph using Plotly Express with custom colors
fig = px.line(
    df,
    x="State",
    y="Median Salary",
    color="Title",
    color_discrete_map=custom_colors,  # Apply custom colors
    labels={"Median Salary": "Mean Salary ($)"},
    title="Average salary by role and by state"
)
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Ensure "Salary" is in numeric format
df['Median Salary'] = pd.to_numeric(df['Median Salary'])

# Create an animated choropleth map using Plotly Express
fig = px.choropleth(
    df,
    locations="State Abbr",
    color="Median Salary",
    locationmode="USA-states",
    scope="usa",
    color_continuous_scale="Viridis",
    labels={"Salary": "Average Salary ($)"},
    title="Average Salary by Role and by State",
    animation_frame="Title",
)

fig.show()
```
